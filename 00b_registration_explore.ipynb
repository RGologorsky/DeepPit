# reorganize by atlas

```python
from helpers.items_constants import *
all_lbl_items = abide_lbl_items + cross_lbl_items
all_lbl_fns   = [i[0] for i in all_lbl_items]

print(len(abide_lbl_items) + len(cross_lbl_items))
```

```python
rows = []
max_n = 0
for atlas,seg in all_lbl_items:

    all_ten = {"atlas_fn": atlas}
    for i in range(10):
        d  = df[df[f"fixed_fn{i}"]==atlas][["moving_fn", f"dice{i}"]]
        d2 = list(zip(d["moving_fn"], d[f"dice{i}"]))

        all_ten.update(
            {
                **{f"dice{i}":     d2[i][1] for i in range(len(d2))},
                **{f"moving_fn{i}":d2[i][0] for i in range(len(d2))}
            }
        )
    # only those who were used as atlas at least once
    if len(all_ten.keys()) > 1:
        rows.append(all_ten)
        
    # append max number
    max_n = max(max_n, len(all_ten.keys()) - 1)
```

```python
print("Max num times used as atlas: ", max_n)
```

```python
import numpy as np
np.NaN
```

```python
# fill in missing cols
def fillin(d, max_n):
    nulld = {
        **{f"dice{i}": np.NaN for i in range(max_n)},
        **{f"moving_fn{i}": np.NaN for i in range(max_n)}
    }
    nulld.update(d)
    return nulld
```

```python
fillin({"dice2": 2}, max_n=3)
```

```python
rows_df = pd.DataFrame([fillin(d, max_n) for d in rows])
rows_df
```

```python
ordered_rows_df = rows_df[["atlas_fn"] + [f"dice{i}" for i in range(10)] + [f"moving_fn{i}" for i in range(10)]]
ordered_rows_df
```

```python
ordered_rows_df['avg'] = ordered_rows_df[[f"dice{i}" for i in range(10)]].mean(axis=1)
sorted_df = ordered_rows_df[["atlas_fn", 'avg']].sort_values("avg", ascending=False)
sorted_df
```
